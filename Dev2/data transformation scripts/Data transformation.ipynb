```python
import pandas as pd

# Step 1: Load both datasets
co2_data = pd.read_csv('co2emissions.csv')
ev_sales_data = pd.read_csv('EV salesHistoricalCars.csv')

# Step 2: Inspect the column names to ensure correct merging
print("CO2 Emissions Data Columns:", co2_data.columns)
print("EV Sales Data Columns:", ev_sales_data.columns)

# Step 3: Rename columns for consistency
# Rename 'Year' in the sales data to match 'Year' in the co2 emissions data
ev_sales_data.rename(columns={'year': 'Year', 'value': 'EV Sales'}, inplace=True)

# Step 4: Merge the two datasets on the 'Year' column
merged_data = pd.merge(co2_data[['Year', 'Total CO2 Emissions Reduction (kg)', 'EV Sales (Units)']],
                       ev_sales_data[['Year', 'EV Sales']], on='Year', how='inner')

# Step 5: Check the merged dataset
print(merged_data.head())

# Step 6: Plot the data
import matplotlib.pyplot as plt

# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot EV Sales on the left y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('EV Sales (Units)', color='tab:blue')
ax1.plot(merged_data['Year'], merged_data['EV Sales'], color='tab:blue', label='EV Sales (Units)')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis to plot CO2 Emissions Reduction
ax2 = ax1.twinx()
ax2.set_ylabel('Total CO2 Emissions Reduction (kg)', color='tab:green')
ax2.plot(merged_data['Year'], merged_data['Total CO2 Emissions Reduction (kg)'], color='tab:green', label='Total CO2 Emissions Reduction')
ax2.tick_params(axis='y', labelcolor='tab:green')

# Add a title
plt.title('EV Sales vs. CO2 Emissions Reduction over Time')

# Show the plot
plt.tight_layout()
plt.show()
```
